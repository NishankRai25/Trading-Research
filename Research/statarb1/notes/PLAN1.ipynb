# Introduction

This paper proposes a **multivariate statistical arbitrage strategy** designed for a universe of highly liquid **AI, technology, and semiconductor equities**, sectors that currently exhibit strong structural growth, high co-movement, and rich relative-value opportunities.

The central hypothesis is that stocks within this universe share common latent risk factors driven by macroeconomic conditions, innovation cycles, semiconductor demand, and market sentiment. Temporary deviations from these common factors may generate **mispricings that mean-revert over time**, creating profitable statistical arbitrage opportunities.

Let $P_t \in \mathbb{R}^N$ denote the vector of log prices for $N$ assets at time $t$. Following the framework of Avellaneda and Lee, returns are decomposed into a systematic factor component and an idiosyncratic residual component using **Principal Component Analysis (PCA)**:

$$
R_t = B F_t + \epsilon_t
$$

where:

- $R_t$ = vector of asset returns  
- $B$ = factor loading matrix  
- $F_t$ = common latent factors (principal components)  
- $\epsilon_t$ = residual spread vector  

The residual $\epsilon_t$ represents the relative mispricing after removing common market and sector movements. This residual forms the core spread traded by the strategy.

---

# Dynamic Spread Modeling

Traditional statistical arbitrage models often assume the spread follows an **Ornstein–Uhlenbeck (OU) process**:

$$
dX_t = \kappa (\mu - X_t)\,dt + \sigma\,dW_t
$$

where:

- $X_t$ = spread / residual process  
- $\mu$ = long-run equilibrium (fair value)  
- $\kappa$ = speed of mean reversion  
- $\sigma$ = volatility  
- $W_t$ = Brownian motion  

A key limitation of the standard OU model is that $\mu$, $\kappa$, and $\sigma$ are usually assumed constant.

This paper instead proposes a **fully time-varying OU framework**:

$$
dX_t = \kappa_t (\mu_t - X_t)\,dt + \sigma_t\,dW_t
$$

where all three parameters evolve dynamically.

---

# Parameter Forecasting Framework

## 1. Dynamic fair value $\mu_t$

The equilibrium spread level is modeled using a **Kalman filter**, allowing fair value to update as new information arrives:

$$
\mu_t = \mu_{t-1} + \eta_t
$$

This captures shifting relative valuations between highly growth-sensitive technology assets.

---

## 2. Regime-dependent mean reversion $\kappa_t$

The speed of reversion is assumed to depend on market regime.

Let latent regime state $S_t$ follow a **Markov chain**:

$$
P(S_t=j \mid S_{t-1}=i)=p_{ij}
$$

Then:

$$
\kappa_t = \kappa(S_t)
$$

This allows mean reversion to behave differently across:

- calm / normal markets  
- high-volatility periods  
- momentum / trend-dominated regimes  

For example, spreads may revert quickly in stable markets but slowly during crisis or narrative-driven rallies.

---

## 3. Time-varying volatility $\sigma_t$

Spread volatility is modeled using econometric and machine learning volatility forecasting techniques.

For example, a **GARCH(1,1)** specification:

$$
\sigma_t^2 = \omega + \alpha \epsilon_{t-1}^2 + \beta \sigma_{t-1}^2
$$

This may later be extended with:

- ARCH / HARCH  
- LSTM / deep learning volatility forecasts  
- hybrid ML-econometric models  

to improve forecast accuracy.

---

# Trading Strategy

Signals are generated from the deviation of the spread from fair value:

$$
z_t = \frac{X_t - \mu_t}{\sigma_t}
$$

Trading rules are regime-specific:

- **long** when $z_t < -z_{entry}(S_t)$
- **short** when $z_t > z_{entry}(S_t)$
- **exit** when $z_t$ reverts toward 0

Position sizing and risk limits will be optimized conditional on regime and forecast volatility.

---

# Contribution

The contribution of this paper is the development of a **dynamic regime-aware multivariate statistical arbitrage framework** that integrates:

- PCA factor residualization
- Kalman-filter fair value estimation
- Markov regime switching
- econometric / deep learning volatility forecasting
- dynamic OU mean reversion

This creates a significantly more flexible framework than classical static spread models.
