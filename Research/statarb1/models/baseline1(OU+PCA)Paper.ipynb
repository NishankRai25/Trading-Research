# Baseline Model I: PCA-Based Statistical Arbitrage with Ornstein–Uhlenbeck Filtering

---

## 1. Introduction and Research Foundation

This study develops a cross-sectional statistical arbitrage strategy grounded in the framework of Avellaneda & Lee (2010), *Statistical Arbitrage in the U.S. Equities Market*. The central premise is that equity returns can be decomposed into systematic and idiosyncratic components, where deviations from common factors represent temporary mispricing opportunities.

Formally, asset returns are modeled as:

$$
R_t = B F_t + \epsilon_t
$$

where $F_t$ denotes latent common factors, $B$ represents factor loadings, and $\epsilon_t$ captures idiosyncratic residuals. These residuals form the basis of the trading signal under the assumption that they exhibit mean-reverting dynamics.

---

## 2. Methodological Adaptations

While the model preserves the core structure of the original framework, several modifications were introduced to improve robustness and realism.

First, residuals are accumulated over time:

$$
X_{i,t} = \sum_{\tau \le t} \epsilon_{i,\tau}
$$

This transformation smooths high-frequency noise and induces dynamics more consistent with Ornstein–Uhlenbeck (OU) behavior.

Second, mean reversion is explicitly enforced via an AR(1)-to-OU mapping:

$$
X_{t+1} = a + bX_t + \zeta_t
$$

$$
\kappa = -\log(b)\cdot252
$$

Only spreads satisfying $\kappa > 14$ are retained, ensuring sufficiently fast mean reversion.

Third, equilibrium levels are centered cross-sectionally:

$$
m_i^{adj} = m_i - \bar{m}
$$

This converts the framework from independent mean-reversion to a relative-value structure.

Finally, execution is made realistic through persistent positions and explicit factor-neutral hedging, removing systematic exposures beyond simple dollar neutrality.

---

## 3. Model Construction

### 3.1 Data and Preprocessing

The model is applied to a universe of approximately 100 highly liquid technology and semiconductor equities. Daily returns are computed as:

$$
r_{i,t} = \frac{P_{i,t}}{P_{i,t-1}} - 1
$$

A rolling window of 252 trading days is used for all estimations.

---

### 3.2 Factor Extraction via PCA

Standardized returns are used to compute a correlation matrix, from which the top $K = 15$ principal components are extracted. These components represent the dominant systematic drivers of the cross-section.

---

### 3.3 Residual Estimation

Each asset’s return is regressed on the factor space:

$$
r_{i,t} = \alpha_i + \beta_i^\top F_t + \epsilon_{i,t}
$$

The residual $\epsilon_{i,t}$ represents the idiosyncratic deviation from common structure.

---

### 3.4 Spread and OU Modeling

Residuals are accumulated into spreads $X_{i,t}$, which are then modeled as OU processes via AR(1) estimation. The equilibrium level and variance are given by:

$$
m = \frac{a}{1-b}
$$

$$
\sigma_{eq} = \sqrt{\frac{\mathrm{Var}(\zeta)}{1-b^2}}
$$

---

### 3.5 Signal Construction

The standardized signal is defined as:

$$
s_i = -\frac{m_i^{adj}}{\sigma_{eq,i}}
$$

This measures the distance of each asset from its equilibrium in standardized units.

---

## 4. Trading Strategy

Positions are determined based on threshold rules:

- Entry: $|s_i| > 1.75$
- Long positions: $s_i < -1.75$
- Short positions: $s_i > 1.75$

Exit rules are asymmetric to account for observed signal decay dynamics.

Portfolio weights are equally distributed across active positions, while factor exposures are neutralized:

$$
w_{factor} = -\sum_i w_i \beta_i
$$

Transaction costs are modeled as:

$$
\text{cost} = c \cdot \sum |w_t - w_{t-1}|
$$

with $c = 5$ basis points.

---

## 5. Empirical Results

The baseline model yields the following performance characteristics:

- Sharpe Ratio: $\approx 0.60$  
- Annual Return: $\approx 5.3\%$  
- Volatility: $\approx 8.9\%$  
- Maximum Drawdown: $\approx -18.6\%$  
- Profit Factor: $\approx 1.11$  
- Average Positions: $\approx 15$  
- Daily Turnover: $\approx 0.35$  

Out-of-sample validation indicates moderate degradation (Sharpe $\approx 0.44$), suggesting limited overfitting.

---

## 6. Parameter Selection and Validation

A structured grid search was conducted over key parameters, including entry thresholds, OU filters, and PCA dimensionality. Model selection was based on Sharpe ratio, followed by out-of-sample validation and local robustness checks to ensure stability across parameter neighborhoods.

---

## 7. Discussion

The results confirm that cross-sectional residual mean reversion is present but relatively weak. While PCA effectively removes systematic structure and OU filtering improves signal quality, several limitations remain.

Signal strength is modest, with limited asymmetry in returns. High turnover introduces significant cost sensitivity, and the use of static equilibrium levels fails to capture evolving market conditions. Additionally, equal weighting ignores cross-sectional heterogeneity in signal strength and risk.

---

## 8. Conclusion

This study demonstrates that a PCA-based residual framework combined with OU filtering produces a functional, factor-neutral statistical arbitrage strategy with stable, though limited, performance.

The findings suggest that while the baseline captures genuine inefficiencies, further improvements—particularly in dynamic fair value estimation, volatility modeling, and regime adaptation—are required to achieve economically significant alpha.
